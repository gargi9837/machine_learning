```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('./Data Visualization Assignmnet.csv')
```

# DATASET CLEANING

```python
df.isnull().sum()     #findinf number of NaN values in the dataset
```

```python
df=df.dropna()         #deleting the rows containing NaN values
df.info()
```

```python
df['PLACE'].value_counts()
```

```python
df['PLACE']=df['PLACE'].str.replace("malappuram","Malappuram")     #combining two same PLACE column only with the difference of upper case and lower case
df['PLACE']=df['PLACE'].str.replace("kadavanthra","Kadavanthra")   #combining two same PLACE column only with the difference of upper case and lower case
df['PLACE']=df['PLACE'].str.replace('Tripunithura ',"Tripunithura")#difference in white spaces, spelling
df['PLACE']=df['PLACE'].str.replace('Tripunithurra',"Tripunithura")#difference in spelling
```

```python
df['DESCRIPTION']=df['DESCRIPTION'].str.replace('Dry Fruits','Dry fruits') ##combining two same PLACE column only with the difference in space in between
```

df['DESCRIPTION'].unique()

```python
df['UNIT']=df['UNIT'].astype(int)                #changing datatype of UNIT to int for graph plotting
```

```python
df.reset_index(drop=True)                        #rearranging index sequence
```

```python
df['DATE']=pd.to_datetime(df['DATE'])             #changing datatype of DATE to DATE for extracting month,day,etc
```

```python
df['MONTH']=df['DATE'].dt.month_name()            #creating MONTH column
```

```python
df['DESCRIPTION'].value_counts().plot(kind='pie',autopct='%1.0f%%')  #highest selling item(DESCRIPTION) is beef is with 29% weightage
```

```python
df['PLACE'].value_counts().plot(kind='bar')      #Malappuram  with the most sales
```

```python
df['MONTH'].value_counts().plot(kind='bar')    #maximum sale is in August
```

```python
df['MONTH'].value_counts().plot(kind='pie',autopct='%1.0f%%')  
```
